Data Discovery with Visualization
==================

In this notebook we will explore how to use python to discover descriptive attributes of various data.

There are thirteen data files contained in the sub-directory `dat/`, with the path and generic basename `dat/dataset_k{:02d}.tsv`. These tab separatated value files contain two columns of floats, `x` and `y`.

We will first explore the data by computing summary statistics.

- - -

```python
# Import required modules
import pandas as pd
import numpy as np
import glob,natsort
# Generate a list of `tsv` files and sort the list
F = natsort.realsorted(glob.glob('dat/dataset_K*.tsv'))
# Generate a list of DataFrames with list index -> dataset index
dfs = [ pd.read_csv(f,delimiter='\t') for f in F ]
```

```python
dfs
```

```python
# Useful function
def gen_summary_statistics(df):
    methods = [ 
        df.mean, 
        df.std, 
        df.skew, 
        df.kurt,
    ]
    return np.r_[
        np.array([ method().values for method in methods ]).flatten(),
        df.corr().values[1,0]
    ]

# Column names of generated summary stats
cols = [
    'x_mean','y_mean',
    'x_std', 'y_std',
    'x_skew','y_skew',
    'x_kurt','y_kurt',
    'corr',
]
```

```python
# Loop through DataFrames and generate summary statistics DataFrame
sdf = pd.DataFrame([gen_summary_statistics(df) for df in dfs ], columns=cols)
sdf
```

```python
# Get to the visualizin
import pylab as plt
import sys
sys.path.append('../common')
from plt_style import *
def mk_grids(): 
    plt.grid(which='major',linestyle='--',color='#CCCCCC')
    plt.grid(which='minor',linestyle=':',color='#AAAAAA')
```

```python
plt.figure(figsize=(8,8))
my_plot(sdf.x_mean,sdf.y_mean,'o',mec=gry,mfc=gld,ms=8); mk_grids()
plt.xlabel('<x>',**lblprops)
plt.ylabel('<y>',**lblprops)
_=plt.title('Mean Comparison',pad=10)
```

```python
plt.figure(figsize=(8,8))
my_plot(sdf.x_std,sdf.y_std,'o',mec=gry,mfc=gld,ms=8); mk_grids()
plt.xlabel('stdev(x)',**lblprops)
plt.ylabel('stdev(y)',**lblprops)
_=plt.title('Standard Dev. Comparison',pad=10)
```

```python
plt.figure(figsize=(8,8))
my_plot(sdf.x_mean/sdf.x_std,sdf.y_mean/sdf.y_std,'o',mec=gry,mfc=gld,ms=8); mk_grids()
plt.xlabel('SNR(x)',**lblprops)
plt.ylabel('SNR(y)',**lblprops)
_=plt.title('Signal to Noise Comparison',pad=10)
```

```python
plt.plot(sdf['corr'],'o',mec='k',mfc=gld,ms=8); mk_grids()
plt.xlabel(r'$k$')
plt.ylabel(r'$\rho(xy)$')
plt.title('Correlation Comparison')
```

```python
plt.plot(sdf['x_skew'],sdf['y_skew'],'o',mec='k',mfc=gld,ms=8); mk_grids()
plt.xlabel(r'$\gamma(x)$')
plt.ylabel(r'$\gamma(y)$')
plt.title('Skewness Comparison')
```

```python
plt.plot(sdf['x_kurt'],sdf['y_kurt'],'o',mec='k',mfc=gld,ms=8); mk_grids()
plt.xlabel(r'$\kappa(x)$')
plt.ylabel(r'$\kappa(y)$')
plt.title('Kurtosis Comparison')
```

What do the data actually look like?
---

```python
f,axarr = plt.subplots(4,4,figsize=(21,20))
# Only 13 plots, remove last 3 empty frames
f.delaxes(axarr[-1,-3])
f.delaxes(axarr[-1,-2])
f.delaxes(axarr[-1,-1])
# loop over dataframes and plot data
for k,df in enumerate(dfs):
    ax = axarr[k//4,k%4]
    ax.plot(df.x,df.y,'o',c='w' if k != 3 else 'r')
    ax.set_xlabel(r'x',**lblprops)
    ax.set_ylabel(r'y',**lblprops)
    ax.set_title(f'Index {k:02d}')
plt.axis('equal')    
plt.tight_layout()
```
