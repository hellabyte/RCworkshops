Exploring Wine Data
===



```python
import pandas as pd
import numpy as np
import pylab as plt
import seaborn as sns
from plt_style import *
%matplotlib inline
sns.set()
```

```python
df = pd.read_csv('dat/winequality-red.csv',delimiter=';')
df
```

```python
df.describe()
```

```python
(df.std() / df.mean()).sort_values()
```

```python
sns.pairplot(df,vars=['citric acid','total sulfur dioxide','density','quality'],hue='quality',palette='husl')
```

```python
df.groupby('quality').nunique()
```

```python
sns.swarmplot(x='quality', y='citric acid',data=df,palette='husl')
```

```python
sns.violinplot(x='quality', y='citric acid',data=df,palette='husl',inner=None)
```
