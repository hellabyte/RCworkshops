Exploring Wine Data
===



```python
import pandas as pd
import numpy as np
import pylab as plt
import seaborn as sns
from plt_style import *
sns.set()
```

```python
df = pd.read_csv('dat/winequality-red.csv',delimiter=';')
df
```

```python
df.describe()
```

```python
(df.std() / df.mean()).sort_values()
```

```python
_ = sns.pairplot(df,vars=['citric acid','total sulfur dioxide','density','quality'],hue='quality',palette='husl')
```

```python
df.groupby('quality').nunique()
```

```python
sns.swarmplot(x='quality', y='citric acid',data=df,palette='husl')
```

```python
sns.violinplot(x='quality', y='citric acid',data=df,palette='husl',inner=None)
```

```python
import plotly.express as px
import plotly.io as pio
import plotly.graph_objects as go
pio.renderers.default = 'iframe'
```

```python
fig=px.scatter_3d(
    df,
    x='citric acid',
    y='density',
    z='quality',
    color='quality',
    color_continuous_scale='viridis',
)
fig.update_traces(
    marker=dict(
        opacity=0.4,
        size=10,
        line=dict(
            width=0,
            color='black',
        ),
    ),
    selector=dict(mode='markers'),
)

fig.update_layout(
    width=1000,
    height=800,
    hoverlabel=dict(
        font=dict(
            family='monospace',
            size=10,
        ),
    ),
    legend_font=dict(color='#eee',size=24),
    legend_itemsizing='constant',
    legend_bgcolor='#444',
    paper_bgcolor='#222',
    plot_bgcolor='#333',
    scene = dict(
        xaxis = dict(
             backgroundcolor="#555",
             gridcolor="#111",
             showbackground=True,
             zerolinecolor="#111",),
        yaxis = dict(
            backgroundcolor="#333",
            gridcolor="#111",
            showbackground=True,
            zerolinecolor="#111"),
        zaxis = dict(
            backgroundcolor="rgba(0,0,0,0)",
            gridcolor="#111",
            showbackground=True,
            zerolinecolor="#111",
        ),
    ),
    font_color="#111",
)


fig.update_layout(paper_bgcolor="#ddd")
```

```python
fig = px.colors.sequential.swatches_continuous()
fig.show()
```
